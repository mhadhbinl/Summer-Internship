```python
test= pd.read_excel ('TEST CHAINE CONTENTIEUX - Copie (1).xlsx')
 test.info()
test.head()
```

```python
#"id "
ids = contract['Lien du contrat'].str.split('/').str[2]
```

```python
#month
month=test['Calendrier'].str.split('/').str[0]
```

```python
contract_test=contract.loc[:,['Date effective','Jour de prélèvement']]
contract_test['ids']=ids
```

```python
contract_test['Date effective']=pd.to_datetime(contract_test['Date effective'])
contract_test['year']= contract_test['Date effective'].dt.year
contract_test['month']= contract_test['Date effective'].dt.month
contract_test['day']= contract_test['Date effective'].dt.day
contract_test.head()
```

```python
chaine_test=test.loc[:,['Jours de prelevement','Calendrier']]
chaine_test['month']=month
#print(chaine_test)
chaine_test['month']=chaine_test['month'].str.upper()

chaine_test.loc[(chaine_test['month']=='JANVIER'), 'month']=1
chaine_test.loc[(chaine_test['month']=='FÉVRIER'), 'month']=2
chaine_test.loc[(chaine_test['month']=='MARS'), 'month']=3
chaine_test.loc[(chaine_test['month']=='AVRIL'), 'month']=4
chaine_test.loc[(chaine_test['month']=='MAI'), 'month']=5
chaine_test.loc[(chaine_test['month']=='JUIN'), 'month']=6
chaine_test.loc[(chaine_test['month']=='JUILLET'), 'month']=7
chaine_test.loc[(chaine_test['month']=='AOÛT'), 'month']=8
chaine_test.loc[(chaine_test['month']=='SEPTEMBRE'), 'month']=9
chaine_test.loc[(chaine_test['month']=='OCTOBRE'), 'month']=10
chaine_test.loc[(chaine_test['month']=='NOVEMBRE'), 'month']=11
chaine_test.loc[(chaine_test['month']=='DÉCEMBRE'), 'month']=12
```

```python
#year
year = chaine_test['Calendrier'].str.split('/').str[1]
chaine_test['year']=year
chaine_test['contrat']=None
chaine_test.head()
```

```python
for i in range(len(chaine_test)):
  j=0
  contract_test=contract_test.reset_index(drop=True)
  while j!= len(contract_test)-1:
    j=j+1
    if int(chaine_test['month'].iloc[i])==int(contract_test['month'].iloc[j]) and
    int(chaine_test['year'].iloc[i])>=int(contract_test['year'].iloc[j]) and int(chaine_test['Jours de prelevement'].iloc[i])==int(contract_test['Jour de prélèvement'].iloc[j]):
      chaine_test['contrat'].iloc[i]=contract_test['ids'].iloc[j]
      contract_test=contract_test.drop(index=j)
      break
```

```python
chaine_test['retour de paiement']=None
for i in range(len(chaine_test)):
  if (chaine_test['contrat'].iloc[i])!=None:
    chaine_test['retour de paiement'].iloc[i]='Oui'
```

```python
import random
chaine_test['date du retour de paiement']=None
chaine_test['date']=None
for i in range(len(chaine_test)):
    chaine_test['date'].iloc[i]=int(chaine_test['Jours de prelevement'].iloc[i])+random.randint(0, 20)
```

```python
chaine_test['date du retour de paiement']=chaine_test['date'].map(str)+'/'+chaine_test['month'].map(str)+'/'+chaine_test['year'].map(str)
for i in range(len(chaine_test)):
  if (chaine_test['contrat'].iloc[i])==None:
    chaine_test['date du retour de paiement'].iloc[i]=None
```

```python
chaine_test['Raison de retour']=None
for i in range(len(chaine_test)):
  if (chaine_test['contrat'].iloc[i])!=None:
    chaine_test['Raison de retour'].iloc[i]='Account closed'
```

```python
test['Contrat']=chaine_test['contrat']
test['retour de paiement']=chaine_test['retour de paiement']
test['date du retour de paiement']=chaine_test['date du retour de paiement']
test['Raison de retour']=chaine_test['Raison de retour']
```
